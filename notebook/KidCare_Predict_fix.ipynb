# **MODEL ARCHITECTURE KIDCARE USING DNN**

```python
print(df['Jenis Kelamin'].unique())
print(df['Status Gizi'].unique())
```

```python
Jenis_Kelamin_encoding = {'Laki Laki': 1, 'Perempuan': 0}
Status_Gizi_encoding = {'tidak stunting': 0, 'stunting': 1}

df['Jenis Kelamin'] = df['Jenis Kelamin'].map(Jenis_Kelamin_encoding)
df['Status Gizi'] = df['Status Gizi'].map(Status_Gizi_encoding)

df.head(6000)
```

```python
# Fitur (X) dan target (y)
X = df[['Jenis Kelamin', 'Umur', 'Berat Badan', 'Tinggi Badan', 'Lingkar Kepala']].values
y = df['Status Gizi'].values

# Cek dimensi data
print("Dimensi X:", X.shape)
print("Dimensi y:", y.shape)
```

```python
from imblearn.over_sampling import SMOTE

# Inisialisasi SMOTE
smote = SMOTE(random_state=42)

# Terapkan SMOTE pada data
X_resampled, y_resampled = smote.fit_resample(X, y)

# Cek dimensi setelah oversampling
print("Dimensi X setelah SMOTE:", X_resampled.shape)
print("Distribusi y setelah SMOTE:", np.bincount(y_resampled))
```

```python
from sklearn.preprocessing import MinMaxScaler
# Inisialisasi scaler
scaler = MinMaxScaler()

# Normalisasi data
X_resampled = scaler.fit_transform(X_resampled)

# Cek data setelah normalisasi
print("Contoh data setelah normalisasi:")
print(X_resampled[:5])
```

```python
from sklearn.model_selection import train_test_split

# Membagi data
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)

# Cek dimensi data setelah split
print("Dimensi X_train:", X_train.shape)
print("Dimensi X_test:", X_test.shape)
```

```python
from sklearn.utils.class_weight import compute_class_weight
import numpy as np

# Ubah daftar menjadi numpy array
classes = np.array([0, 1])

# Hitung bobot kelas
class_weights = compute_class_weight(class_weight='balanced', classes=classes, y=df['Status Gizi'].values)
print("Class Weights:", class_weights)
```

```python
from tensorflow.keras.callbacks import EarlyStopping

# Definisikan EarlyStopping callback
early_stopping = EarlyStopping(monitor='val_loss',  # Atau 'val_accuracy' jika ingin memonitor akurasi
                               patience=10,         # Menunggu 10 epoch tanpa perbaikan
                               restore_best_weights=True)

# Definisikan model
model = Sequential([
    Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')  # Aktivasi sigmoid untuk binary classification
])

# Kompilasi model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Latih model dengan callback EarlyStopping
history = model.fit(X_train, y_train, class_weight={0: 1.36492221, 1: 0.78904334}, validation_data=(X_test, y_test), epochs=1000, batch_size=32, callbacks=[early_stopping])
```

```python
#grafik check train n test
plt.plot(model.history.history['accuracy'])
plt.plot(model.history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()

plt.plot(model.history.history['loss'])
plt.plot(model.history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
```

# **EVALUASI MODEL**

```python
# Make predictions on the test set
y_pred = model.predict(X_test)

# Convert continuous variable to binary labels
threshold = 0.5
y_pred_binary = (y_pred >= threshold).astype(int)

# Create the confusion matrix
confusion = confusion_matrix(y_pred_binary, y_test)

# Create a heatmap of the confusion matrix
sns.heatmap(confusion, annot=True, cmap="Blues", fmt="d")

# Set the axis labels
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")

# Set the title
plt.title("Confusion Matrix")

# Display the plot
plt.show()
```

```python
# Evaluasi model pada data pengujian
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f'Test Accuracy: {test_accuracy:.4f}')
```

```python
# Calculate precision, recall, and F1 score
precision = precision_score(y_test, y_pred_binary)
recall = recall_score(y_test, y_pred_binary)
f1 = f1_score(y_test, y_pred_binary)

# Generate classification report
report = classification_report(y_test, y_pred_binary)

# Print the metrics report
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
print("Classification Report:\n", report)
```

```python
# Definisikan threshold untuk klasifikasi
threshold = 0.5

# Prediksi untuk input pertama
prediction1 = model.predict(np.array([[26, 1, 6, 86, 31]]))
print(f"Prediksi untuk input pertama: {prediction1}")

# Menentukan klasifikasi berdasarkan threshold
if prediction1[0] > threshold:
    print("Status Gizi: Stunting")
else:
    print("Status Gizi: Tidak Stunting")

# Prediksi untuk input kedua
prediction2 = model.predict(np.array([[26, 0, 12, 98, 37]]))
print(f"Prediksi untuk input kedua: {prediction2}")

# Menentukan klasifikasi berdasarkan threshold
if prediction2[0] > threshold:
    print("Status Gizi: Stunting")
else:
    print("Status Gizi: Tidak Stunting")
```
